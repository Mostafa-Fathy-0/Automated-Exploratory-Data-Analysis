```python
# Automated EDA tool

# importation of libraries
import pandas as pd
import matplotlib.pyplot as plt

# get_path()
# take path from user and check if it is in range of us or not
# our paths range [csv , xlsx , sql]
def get_path(file_type) :
    while True:
        file_path = input("Please enter the file path: ")
        if file_path.endswith(tuple(file_type)):
            return file_path
        print("Invalid file path.Please enter a path from this range: {}".format(", ".join(file_type)))

# visualizations()
# this function deal with every column and plot it with a most fit chart for it's data type
# categorical data ----> bar chart
# numerical data ----> histogram chart
def visualizations(df):

    # determine type of each column to choose best way to plot it
    for column_name in df.columns:
        column_data = df[column_name]
        column_type = column_data.dtype

        # create a plots for the column.
        fig, axs = plt.subplots()

        # make visualization for all data types.

        # categorical data
        # we will visualize it by using bar chart
        if column_type == 'object':
            value_counts = column_data.value_counts()
            labels = value_counts.index
            colors = ['cadetblue', 'coral', 'limegreen', 'mediumpurple']
            axs.bar(labels, value_counts, color=colors)

        # numerical data
        # we will visualize it by using hist chart
        elif column_type in ['int64', 'float64']:
            axs.hist(column_data, bins='auto', color='skyblue')
            axs.axvline(column_data.mean(), color='red', linestyle='dashed', linewidth=1)
            axs.axvline(column_data.median(), color='green', linestyle='dashed', linewidth=1)
            axs.legend(['Mean', 'Median'])
    # set title for each column
        axs.set_title(column_name)
    # grid axis
        axs.grid(True)
    # show visualization
        plt.show()

# main()
# this is where our code runs
def main():

    # our paths range for files
    valid_path = ['.csv', '.xlsx', '.sql']

    # prompt the user to enter a valid file path.
    file_path = get_path(valid_path)

    # load the data files based on the file path.
    # deal with csv files
    if file_path.endswith('.csv'):
        df = pd.read_csv(file_path.strip())
    # deal with excel files
    elif file_path.endswith('.xlsx'):
        df = pd.read_excel(file_path.strip())
    # deal with sql data bases
    elif file_path.endswith('.sql'):
        # i have made search for many hours but i didn't understand how to deal with sql base.
        # so  i pass it right now.
        # but i will try again to complete this part and know how to handle sql database
        print("Sorry!,Handle sql databases is not available righ now")

    # calling visualizations functions of the data.
    visualizations(df)

# call the main function to run the tool
if __name__ == "__main__":
    main()
```
